```python
import pandas as pd
```

```python
#Set up variables for filepaths
cryptofilepath = "crypto_data.csv"
```

```python
#pull file into dataframe

crypto_df = pd.read_csv(cryptofilepath, encoding="UTF-8")
```

```python
#Review first 5 records of file
crypto_df.head()                 #1252 Records
#crypto_df.count()               #1252 Records, only 744 Mined - so delta is 508. 
```

```python
#Find records of that have isTrading =True
TradingCrypto_df=crypto_df.loc[crypto_df['IsTrading'] == True]

TradingCrypto_df.head()        #1144 Records
#TradingCrypto_df.count()        #685 Records
```

```python
#Remove IsTrading Column
del TradingCrypto_df['IsTrading']
```

```python
# Checkscript
TradingCrypto_df.head()
#TradingCrypto_df.count()       #1144, 685 
```

```python
TradingCrypto_df = TradingCrypto_df.dropna(how='any',axis=0)
```

```python
# Checkscript
TradingCrypto_df.head()
#TradingCrypto_df.count()      #685 Records across all 
```

```python
#Identify records that are not unmined
TradingMinedCrypto_df = TradingCrypto_df[~TradingCrypto_df['TotalCoinsMined'].isnull()]
TradingMinedCrypto_df.head()        #685 Total Records
#TradingMinedCrypto_df.count()      #685 Total Records
```

```python
#TradingMinedCrypto_df.head()

#Re-define datatype needed
TradingMinedCrypto_df["TotalCoinSupply"] = pd.to_numeric(TradingMinedCrypto_df["TotalCoinSupply"])
#TradingMinedCrypto_df.dtypes

#Find records of that have isTrading =True
TradingMinedNonZeroCrypto_df=TradingMinedCrypto_df.loc[(TradingMinedCrypto_df['TotalCoinSupply'] > 0)]

TradingMinedNonZeroCrypto_df.head()        #637 Records
#TradingMinedNonZeroCrypto_df.count()        #637 Records
```

```python
#Checkscript
TradingMinedNonZeroCrypto_df.head() 
```

```python
# Convert categorical data to numeric by 
# One-hot encoding the entire dataframe
CryptoOneHot_df = pd.get_dummies(TradingMinedNonZeroCrypto_df)
print(CryptoOneHot_df.columns)
CryptoOneHot_df
```

This step set a multitude of binary columns in place of the named columns. We went from 6 columns to 1378 columns. 

```python
#Remove CoinName Column from the dataframe
#del TradingMinedNonZeroCrypto_df['CoinName']
TradingMinedNonZeroCrypto_df.head()
```

```python
# Scale the data in the fields that have large values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(CryptoOneHot_df)
```

```python
# Create a DataFrame with the transformed data
#FinalCrypto_df = pd.DataFrame(scaled_data)

# Create a DataFrame with the transformed data
FinalCrypto_df = pd.DataFrame(scaled_data, columns=CryptoOneHot_df.columns[:])

FinalCrypto_df.head()
#FinalCrypto_df.head()
```
